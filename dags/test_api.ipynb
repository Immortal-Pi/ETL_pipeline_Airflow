```python
# data['Time Series (Digital Currency Daily)'][0]
import psycopg2

url='https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=USD&apikey=F2UL8228YNAAADKA'
conn=psycopg2.connect(
            database="postgres",
            user="postgres",
            password="postgres",
            host="127.0.0.1",
            port=5432,
        )
r=requests.get(url)
data=r.json()
data
dates = list(data['Time Series (Digital Currency Daily)'].keys())
curr=conn.cursor()
insert_query=""" 
        insert into bitcoin_data (title,Digital_Currency_Code,Digital_Currency_Name,Market_Code,Market_Name,Last_Refreshed_Date,Time_Zone,open_price,high_price,low_price,close_price)
        values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
        """
for date in dates:
    btc_data = {
        'title': data['Meta Data']['1. Information'],
        'Digital_Currency_Code': data['Meta Data']['2. Digital Currency Code'],
        'Digital_Currency_Name': data['Meta Data']['3. Digital Currency Name'],
        'Market_Code': data['Meta Data']['4. Market Code'],
        'Market_Name': data['Meta Data']['5. Market Name'],
        'Last_Refreshed_Date': date,
        'Time_Zone': data['Meta Data']['7. Time Zone'],
        'open_price': data['Time Series (Digital Currency Daily)'][date]['1. open'],
        'high_price': data['Time Series (Digital Currency Daily)'][date]['2. high'],
        'low_price': data['Time Series (Digital Currency Daily)'][date]['3. low'],
        'close_price': data['Time Series (Digital Currency Daily)'][date]['4. close'],  # Extract only close price
    }
    
    curr.execute(insert_query,(
            btc_data['title'],
            btc_data['Digital_Currency_Code'],
            btc_data['Digital_Currency_Name'],
            btc_data['Market_Code'],
            btc_data['Market_Name'],
            btc_data['Last_Refreshed_Date'],
            btc_data['Time_Zone'],
            btc_data['open_price'],
            btc_data['high_price'],
            btc_data['low_price'],
            btc_data['close_price'],
        ))
conn.commit()

    
```
